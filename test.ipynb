```python
import time
from pettingzoo.mpe import simple_spread_v3
env = simple_spread_v3.env(render_mode='human', N=3, local_ratio=0.5, max_cycles=250, continuous_actions=False)

# make 4 landmarks in the environment

env.reset()
for agent in env.agent_iter():
    observation, reward, termination, truncation, info = env.last()

    if termination or truncation:
        action = None
    else:
        action = env.action_space(agent).sample() # this is where you would insert your policy

    env.step(action)
    # pause for 0.1 seconds
    time.sleep(0.05)

env.close()
```

```python
import numpy as np
import gzip
import pickle
import time

# Create an example NumPy array with random floats
array = np.random.rand(500000, 3)

# Record the start time
start_time = time.time()

# Serialize and compress the array
compressed_array = gzip.compress(pickle.dumps(array))

# Record the end time
end_time = time.time()

# Calculate the time taken to compress
compression_time = end_time - start_time

print(f"Time taken to compress: {compression_time:.2f} seconds")
```
